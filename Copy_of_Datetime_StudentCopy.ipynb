```python
import datetime as dt
```

**datetime.now():**

It returns the current time in Greenwich Mean Time (GMT) or Coordinated Universal Time (UTC) time zone.

```python
current=dt.datetime.now()
print(current)
```

**Note:** The default format of the `datetime` object is `YYYY-MM-DD HH:MM:SS`

```python
type(current)
```

```python
current.day
```

the `day, month, year, hour, minute` and `second` attributes are used to get the corresponding day, month, year, hour, minute and second values.

```python
print(f"The current year is {current.year}")
print(f"The current month is {current.month}")
print(f"The current day is {current.day}")
print(f"The current hour is {current.hour}")
print(f"The current minute is {current.minute}")
print(f"The current second is {current.second}")
```

```python
type(current.month)
```

**timedelta:**

often used to calculate differences between dates or times.

**Note:** IST= UTC+5:30

```python
ist=dt.datetime.now()+dt.timedelta(hours=5,minutes=30)
print(ist)
```

```python
type(ist)
```

```python
us=dt.datetime.now()-dt.timedelta(hours=7)
print(us)
```

```python
d=ist.date()
print(d)
type(d)
```

```python
t=ist.time()
print(t)
type(t)
```

**strftime()` & `strptime()`**

**strftime():**

convert the date and time values into a string in the desired format.


[The `strftime()` & `strptime()` Format Codes](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior)

**`strptime():`**

We can create a `datetime` object by passing the string values to the `strptime()` function.

We can create separate `datetime.date` and `datetime.time` objects by passing values to the `date()` and `time()` functions.

**to_datetime**:

Converts various date and time representations into pandas datetime objects.
